```python
# Select specific columns from the DataFrame 'df' and create a new DataFrame
df = df[['coaster_name', 
    # 'Length', 'Speed', 
    'Location', 'Status', 
    # 'Opening date','Type', 
    'Manufacturer', 
    # 'Height restriction', 'Model', 'Height','Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section','Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
    # 'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced','Track layout', 'Fastrack available', 'Soft opening date.1','Closing date', 'Opened', 
    # 'Replaced by', 'Website','Flash Pass Available', 'Must transfer from wheelchair', 'Theme','Single rider line available', 'Restraint Style','Flash Pass available', 'Acceleration', 'Restraints', 'Name',
    'year_introduced', 'latitude', 'longitude', 'Type_Main','opening_date_clean', 
    # 'speed1', 'speed2', 'speed1_value', 'speed1_unit',
    'speed_mph', 
    # 'height_value', 'height_unit', 
    'height_ft',
    'Inversions_clean', 'Gforce_clean']].copy()
```

```python
# Retrieve the dimensions (number of rows and columns) of the DataFrame 'df'
df.shape
```

```python
# Retrieve the data types of each column in the DataFrame 'df'
df.dtypes
```

```python
# Convert the 'opening_date_clean' column to datetime data type
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
# Rename selected columns in the DataFrame 'df'
df = df.rename(columns={'coaster_name':'Coaster_Name', 
                    'year_introduced':'Year_Introduced', 
                    'latitude':'Latitude', 
                    'longitude':'Longitude', 
                    'opening_date_clean':'Opening_Date', 
                    'speed_mph':'Speed_Mph', 
                    'height_ft':'Height_Ft', 
                    'Inversions_clean':'Inversions', 
                    'Gforce_clean':'Gforce'})
```

```python
# Display the first few rows of the DataFrame 'df'
df.head()
```

```python
# Calculate the number of missing values in each column of the DataFrame 'df'
df.isna().sum()
```

```python
# Retrieve the duplicate rows in the DataFrame 'df'
df.loc[df.duplicated()]
```

```python
# Retrieve the duplicate rows in the DataFrame 'df' based on the 'Coaster_Name' column
df.loc[df.duplicated(subset='Coaster_Name')]
```

```python
# Retrieve the rows in the DataFrame 'df' where the 'Coaster_Name' is "Crystal Beach Cyclone"
df.query('Coaster_Name == "Crystal Beach Cyclone"')
```

```python
# Retrieve the column names of the DataFrame 'df'
df.columns
```

```python
# Remove duplicate rows in the DataFrame 'df' based on 'Coaster_Name', 'Location', and 'Opening_Date' columns
df = df.loc[~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date'])].reset_index(drop=True).copy()
```

```python
# Retrieve the dimensions (number of rows and columns) of the DataFrame 'df'
df.shape
```

```python
# Create a bar plot showing the top 10 years when coasters were introduced
ax = df['Year_Introduced'].value_counts().head(10).plot(kind='bar', 
                                                   title='Top 10 Years Coasters Introduced')
# Set the x-axis label
ax.set_xlabel('Year Introduced')
# Set the y-axis label
ax.set_ylabel('Count')
```

```python
# Create a histogram plot of coaster speeds in mph
ax = df['Speed_Mph'].plot(kind='hist', 
                     bins=20, 
                     title='Coaster Speed (mph)')
# Set the x-axis label
ax.set_xlabel('Speed (mph)')
```

```python
# Create a kernel density estimation (KDE) plot of coaster speeds in mph
ax = df['Speed_Mph'].plot(kind='kde', 
                     title='Coaster Speed (mph)')
# Set the x-axis label
ax.set_xlabel('Speed (mph)')
```

```python
# Create a scatter plot of coaster speed vs. height
df.plot(kind='scatter', x='Speed_Mph', y='Height_Ft', title='Coaster Speed vs. Height')
# Display the plot
plt.show()
```

```python
# Create a scatter plot of coaster speed vs. height, colored by year of introduction
sns.scatterplot(x='Speed_Mph', y='Height_Ft', data=df, hue='Year_Introduced')
# Display the plot
plt.show()
```

```python
# Create a pair plot of selected variables, with hue by 'Type_Main'
sns.pairplot(df, vars=['Year_Introduced', 'Speed_Mph', 'Height_Ft', 'Inversions', 'Gforce'], 
             hue='Type_Main')
# Display the plot
plt.show()
```

```python
# Calculate the correlation matrix for selected columns, after dropping any rows with missing values
df_corr = df[['Year_Introduced', 'Speed_Mph', 'Height_Ft', 'Inversions', 'Gforce']].dropna().corr()
```

```python
# Create a heatmap of the correlation matrix with annotations
sns.heatmap(df_corr, annot=True)
```

```python
# Create a horizontal bar plot of average coaster speed by location
ax = df.query('Location != "Other"').groupby('Location')['Speed_Mph'].agg(['mean', 'count']).query('count >= 10').sort_values('mean')['mean'].plot(kind='barh', 
                                                                                                                                              figsize=(12, 5), 
                                                                                                                                              title='Average Coast Speed by Location')
# Set the x-axis label
ax.set_xlabel('Average Coaster Speed')
# Display the plot
plt.show()
```
